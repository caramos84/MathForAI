**1.1. Lógica proposicional**

Una proposición es una afirmación que puede ser verdadera o falsa. Ejemplo:

"Está lloviendo" → puede ser True o False.

Operadores básicos:

- and (∧)

- or (∨)

- not (¬)

- => (implicación)

- <=> (bicondicional)

```python
a = True
b = False

print("a y b", a and b)
print("a o b", a or b)
print("no a", not a)
```

**1.2. Teoría de conjuntos**

Un conjunto es simplemente una colección de objetos. Ejemplo:


```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print("Unión de A y B", A.union(B))
print("Intersección de A y B", A.intersection(B))
print("Diferencia de A y B", A.difference(B))
```

Digramas de Venn

```python
# Reimportamos librerías y volvemos a ejecutar los conjuntos y operaciones
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Definimos los conjuntos
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Imprimimos operaciones básicas de teoría de conjuntos
union = A.union(B)
interseccion = A.intersection(B)
diferencia = A.difference(B)

union, interseccion, diferencia

# Crear el diagrama de Venn para los conjuntos A y B
plt.figure(figsize=(4, 4))
venn2([A, B], set_labels=('A', 'B'))
plt.title("Diagrama de Venn de A y B")
plt.show()
```

**1.3. Relaciones (conectar los puntos)**

Una relación es cualquier regla que conecta elementos de un conjunto con otro.

Ejemplo: relación de "es divisible por x número"


```python
A = list(range(1, 10))
relation = [(a, b) for a in A for b in A if b % a == 0]

print("Relación de divisibilidad:")
for pair in relation:
    print(pair)
```

Grafos de la relación

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definir el conjunto y la relación
A = list(range(1, 10))
relation = [(a, b) for a in A for b in A if b % a == 0]

# Crear un grafo dirigido
G = nx.DiGraph()
G.add_edges_from(relation)

# Dibujar el grafo
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=300, arrowstyle='->', arrowsize=10)
nx.draw_networkx_edges(G, pos, edgelist=relation, edge_color='gray')
plt.title("Grafo dirigido: cada nodo apunta a sus múltiplos")
plt.axis('off')
plt.show()
```

**1.4. Vectores**

1.4.1. Definición Matemática

Un vector es un objeto matemático que tiene:

- Magnitud (longitud)

- Dirección

En notación, un vector en 2 dimensiones puede representarse como:

$$
\vec{v} = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}
$$

Donde:

- *v1* es la componente en el eje *x*

- *v2* es la componente en el eje *y*

1.4.2. Operaciones Básicas con Vectores

- Suma de vectores:

$$
\vec{a} + \vec{b} = \begin{bmatrix} a_1 \\ a_2 \end{bmatrix} + \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \end{bmatrix}
$$

- Multiplicación por escalar:

$$
k \cdot \vec{a} = \begin{bmatrix} k a_1 \\ k a_2 \end{bmatrix}
$$

- Magnitud (norma):

$$
|\vec{a}| = \sqrt{a_1^2 + a_2^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir vectores
a = np.array([2, 3])
b = np.array([1, -1])
k = 2 #Escalar

# Suma
suma = a + b
print("Suma:", suma)

# Multiplicación por escalar
multiplicacion = k * a
print("Multiplicación por escalar:", multiplicacion)

# Magnitud
mag_a = np.linalg.norm(a)
print("Magnitud de a:", mag_a)

# Visualización
plt.figure(figsize=(5, 5))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b')
plt.quiver(0, 0, suma[0], suma[1], angles='xy', scale_units='xy', scale=1, color='green', label='a + b')
plt.quiver(0, 0, multiplicacion[0], multiplicacion[1], angles='xy', scale_units='xy', scale=1, color='orange', label='k * a')

plt.xlim(-2, 6)
plt.ylim(-2, 6)
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("Operaciones con Vectores en 2D")
plt.show()
```
